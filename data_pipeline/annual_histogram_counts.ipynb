# Annual Histogram Counts

Load `output/annual_histogram_counts.csv` and render an Altair bar chart with a maroon palette and compact year/count labels.

```python
from pathlib import Path

import altair as alt
import pandas as pd

alt.data_transformers.disable_max_rows()

ROOT = Path.cwd().parent if Path.cwd().name == "data_pipeline" else Path.cwd()
CSV_PATH = ROOT / "output" / "annual_histogram_counts.csv"

df = pd.read_csv(CSV_PATH)
df["year"] = df["year"].astype(int)
df["count"] = df["count"].astype(int)
df["year_label"] = df["year"].astype(str)
year_tick_values = df.loc[df["year"] % 10 == 0, "year_label"].tolist()
y_tick_values = list(range(0, int(df["count"].max()) + 51, 50))

maroon = "#7A0019"
maroon_dark = "#5A0013"
cream = "#eaeaea"
ink = "#2A1A1F"
grid = "#D8C7C9"

base = alt.Chart(df).encode(
    x=alt.X(
        "year_label:N",
        sort=df["year_label"].tolist(),
        title=None,
        axis=alt.Axis(
            values=year_tick_values,
            labelAngle=0,
            labelColor=ink,
            tickColor=grid,
            domainColor=grid,
            labelFontSize=16,
        ),
    ),
    y=alt.Y(
        "count:Q",
        title="Documents",
        axis=alt.Axis(
            values=y_tick_values,
            format="~s",
            labelColor=ink,
            titleColor=ink,
            gridColor=grid,
            domain=False,
            labelFontSize=16,
            titleFontSize=18,
        ),
    ),
    tooltip=[
        alt.Tooltip("year:O", title="Year"),
        alt.Tooltip("count:Q", title="Count", format=",")
    ],
)

bars = base.mark_bar(
    color=maroon,
    size=10,
    cornerRadiusTopLeft=2,
    cornerRadiusTopRight=2,
).encode(
    color=alt.condition(
        alt.datum.count == df["count"].max(),
        alt.value(maroon_dark),
        alt.value(maroon),
    )
)


chart = (bars).properties(
    width=920,
    height=300,
    title=alt.TitleParams(
        text="Number of archival documents by year",
        anchor="middle",
        color=ink,
        fontSize=22,
    ),
).configure_view(
    stroke=None,
    fill=cream,
).configure_axis(
    labelFontSize=16,
    titleFontSize=18,
).configure_title(
    fontSize=22,
).configure(background=cream)

chart
```
