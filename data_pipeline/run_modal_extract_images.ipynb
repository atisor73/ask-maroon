# Run Modal Image Extraction

This notebook is the launchpad for distributed image extraction on Modal.
It does not contain the extraction logic itself. Instead, it calls
[`modal_extract_images.py`](./modal_extract_images.py), which imports the shared
pipeline utilities and runs the remote workers.

Use this notebook to:
- verify local setup
- preview the exact `modal run` command
- launch a small test batch
- then inspect results with `test_image_extraction.ipynb`


## Requirements

Before running:

```bash
conda env create -f img_extractor.yml
conda activate image-extractor
```

You also need:
- the Modal CLI authenticated locally
- a Modal secret named `cloudflare-r2` by default
- R2 credentials stored in that Modal secret using these keys:
  - `R2_ACCOUNT_ID`
  - `R2_ACCESS_KEY_ID`
  - `R2_SECRET_ACCESS_KEY`
  - `R2_BUCKET`
  - optional `R2_ENDPOINT_URL`


````python
from pathlib import Path
import shlex
import subprocess

project_root = Path("..").resolve()
runner_path = Path("modal_extract_images.py").resolve()

print("Project root:", project_root)
print("Modal runner:", runner_path)
````

## Configure A Run

Start small. A good first smoke test is 1-5 PDFs with batch size 1 or 2.

````python
limit = 2
batch_size = 1
page_limit = 0  # use 0 for all pages, or a small integer like 2 for quicker tests
dpi = 144
force = False
use_gpu = True
r2_prefix = "archive"
score_threshold = 0.70
padding = 8
min_side = 48
````

````python
command_parts = [
    "modal",
    "run",
    str(runner_path),
    "--limit",
    str(limit),
    "--batch-size",
    str(batch_size),
    "--r2-prefix",
    r2_prefix,
    "--dpi",
    str(dpi),
    "--score-threshold",
    str(score_threshold),
    "--padding",
    str(padding),
    "--min-side",
    str(min_side),
]

if use_gpu:
    command_parts.append("--use-gpu")
if page_limit > 0:
    command_parts.extend(["--page-limit", str(page_limit)])
if force:
    command_parts.append("--force")

command_string = " ".join(shlex.quote(part) for part in command_parts)
print(command_string)
````

## Launch The Modal Job

Uncomment the last line when you are ready to actually run it.

````python
print("About to run:")
print(command_string)

# result = subprocess.run(command_parts, cwd=project_root, check=True, text=True)
# print("Modal run finished with return code", result.returncode)
````
